```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score
from tensorflow.keras.applications import MobileNet, DenseNet121
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.layers import GlobalAveragePooling2D
from imblearn.over_sampling import SMOTE
from collections import Counter

# === Parameters ===
dataset_path = "/content/drive/MyDrive/UCMerced_LandUse/UCMerced_LandUse/Images"
img_size = (128, 128)
input_shape = img_size + (3,)
n_splits = 5
max_images_per_class = None  # Set to None to use full dataset or e.g., 50 for quicker runs

# === Load Data ===
data, labels = [], []
categories = sorted(os.listdir(dataset_path))
num_classes = len(categories)
print(f"Detected {num_classes} classes.")

for label_index, category in enumerate(categories):
    category_path = os.path.join(dataset_path, category)
    image_files = os.listdir(category_path)
    if max_images_per_class:
        image_files = image_files[:max_images_per_class]

    for filename in image_files:
        img_path = os.path.join(category_path, filename)
        try:
            img = load_img(img_path, target_size=img_size)
            img_array = img_to_array(img)
            data.append(img_array)
            labels.append(label_index)
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")

data = np.array(data, dtype='float32') / 255.0
labels = np.array(labels)

# === Load Pre-trained Models with Global Pooling ===
mobilenet_base = MobileNet(include_top=False, weights="imagenet", input_shape=input_shape)
densenet_base = DenseNet121(include_top=False, weights="imagenet", input_shape=input_shape)
mobilenet_base.trainable = False
densenet_base.trainable = False

mobilenet_model = tf.keras.Sequential([mobilenet_base, GlobalAveragePooling2D()])
densenet_model = tf.keras.Sequential([densenet_base, GlobalAveragePooling2D()])

# === Feature Extraction in Batches ===
def extract_features_batch(model, data, batch_size=32):
    dataset = tf.data.Dataset.from_tensor_slices(data).batch(batch_size)
    features = []
    for batch in dataset:
        feature_batch = model(batch, training=False).numpy()
        features.append(feature_batch)
    return np.concatenate(features, axis=0)

print("Extracting MobileNet features...")
mobilenet_features = extract_features_batch(mobilenet_model, data)

print("Extracting DenseNet features...")
densenet_features = extract_features_batch(densenet_model, data)

# === Combine Features ===
X_features = np.concatenate([mobilenet_features, densenet_features], axis=1)
y_labels = labels

# === K-Fold Cross-Validation with SMOTE + SVM ===
kfold = KFold(n_splits=n_splits, shuffle=True, random_state=42)
fold_accuracies, fold_precisions, fold_recalls = [], [], []

for fold, (train_idx, test_idx) in enumerate(kfold.split(X_features)):
    print(f"\n=== Fold {fold + 1} ===")
    X_train, X_test = X_features[train_idx], X_features[test_idx]
    y_train, y_test = y_labels[train_idx], y_labels[test_idx]

    print("Before SMOTE:", Counter(y_train))
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
    print("After SMOTE:", Counter(y_train_resampled))

    # Train SVM
    svm = SVC(kernel='linear', probability=True)
    svm.fit(X_train_resampled, y_train_resampled)

    # Predict
    y_pred = svm.predict(X_test)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_test, y_pred, average='macro', zero_division=0)

    fold_accuracies.append(acc)
    fold_precisions.append(precision)
    fold_recalls.append(recall)

    print(f"Accuracy: {acc:.4f}")
    print(f"Precision (macro): {precision:.4f}")
    print(f"Recall (macro): {recall:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=categories))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=categories, yticklabels=categories)
    plt.title(f'Fold {fold + 1} - Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()

# === Plot Metrics Across Folds ===
folds = np.arange(1, n_splits + 1)
plt.figure(figsize=(12, 6))
plt.plot(folds, fold_accuracies, marker='o', label='Accuracy')
plt.plot(folds, fold_precisions, marker='o', label='Precision (macro)')
plt.plot(folds, fold_recalls, marker='o', label='Recall (macro)')
plt.title('Evaluation Metrics Across Folds')
plt.xlabel('Fold')
plt.ylabel('Score')
plt.ylim(0, 1.05)
plt.xticks(folds)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score
from tensorflow.keras.applications import MobileNet, DenseNet121
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from tensorflow.keras.layers import GlobalAveragePooling2D, MaxPooling2D, Flatten
from imblearn.over_sampling import SMOTE
from collections import Counter

# === Parameters ===
dataset_path = "/content/drive/MyDrive/UCMerced_LandUse/UCMerced_LandUse/Images"
img_size = (128, 128)
input_shape = img_size + (3,)
n_splits = 5

# === Augmentation Setup ===
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

# === Load and Augment Data ===
data, labels = [], []
categories = sorted(os.listdir(dataset_path))
num_classes = len(categories)
print(f"Detected {num_classes} classes.")

for label_index, category in enumerate(categories):
    category_path = os.path.join(dataset_path, category)
    for filename in os.listdir(category_path):
        img_path = os.path.join(category_path, filename)
        try:
            img = load_img(img_path, target_size=img_size)
            img_array = img_to_array(img)
            data.append(img_array)
            labels.append(label_index)
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")

# Convert to numpy arrays
data = np.array(data, dtype='float32') / 255.0
labels = np.array(labels)

# === Load Pre-trained Models ===
mobilenet_base = MobileNet(include_top=False, weights="imagenet", input_shape=input_shape)
densenet_base = DenseNet121(include_top=False, weights="imagenet", input_shape=input_shape)
mobilenet_base.trainable = False
densenet_base.trainable = False

# === Efficient Feature Extraction (Batched) ===
def extract_features(data, batch_size=32):
    features = []
    for start in range(0, len(data), batch_size):
        batch_data = data[start:start + batch_size]
        batch_data_expanded = np.expand_dims(batch_data, axis=-1)

        mobilenet_feat = mobilenet_base(batch_data_expanded, training=False)
        mobilenet_feat = MaxPooling2D(pool_size=(2, 2))(mobilenet_feat)
        mobilenet_feat = Flatten()(mobilenet_feat)

        densenet_feat = densenet_base(batch_data_expanded, training=False)
        densenet_feat = GlobalAveragePooling2D()(densenet_feat)

        combined_feat = tf.concat([mobilenet_feat, densenet_feat], axis=-1).numpy()
        features.append(combined_feat)

    return np.concatenate(features, axis=0)

print("Extracting features...")
X_features = extract_features(data)
y_labels = labels

# === K-Fold Cross-Validation with SMOTE + SVM ===
kfold = KFold(n_splits=n_splits, shuffle=True, random_state=42)
fold_accuracies = []
fold_precisions = []
fold_recalls = []
losses = []  # Track loss (inverse of accuracy)

for fold, (train_idx, test_idx) in enumerate(kfold.split(X_features)):
    print(f"\n=== Fold {fold + 1} ===")
    X_train, X_test = X_features[train_idx], X_features[test_idx]
    y_train, y_test = y_labels[train_idx], y_labels[test_idx]

    print("Before SMOTE:", Counter(y_train))
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
    print("After SMOTE:", Counter(y_train_resampled))

    # Train SVM
    svm = SVC(kernel='linear', probability=True)
    svm.fit(X_train_resampled, y_train_resampled)

    # Predict
    y_pred = svm.predict(X_test)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_test, y_pred, average='macro', zero_division=0)

    fold_accuracies.append(acc)
    fold_precisions.append(precision)
    fold_recalls.append(recall)
    losses.append(1 - acc)  # Loss is inverse of accuracy for visualization

    print(f"Accuracy: {acc:.4f}")
    print(f"Precision (macro): {precision:.4f}")
    print(f"Recall (macro): {recall:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=categories))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=categories, yticklabels=categories)
    plt.title(f'Fold {fold + 1} - Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()

# === Plotting Metrics Across Folds ===
folds = np.arange(1, n_splits + 1)

# Accuracy and Loss Graphs
plt.figure(figsize=(12, 6))
plt.plot(folds, fold_accuracies, marker='o', label='Accuracy')
plt.plot(folds, losses, marker='o', label='Loss (1 - Accuracy)', linestyle='--')
plt.title('Accuracy and Loss Across Folds')
plt.xlabel('Fold')
plt.ylabel('Score')
plt.ylim(0, 1.05)
plt.xticks(folds)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot Precision and Recall
plt.figure(figsize=(12, 6))
plt.plot(folds, fold_precisions, marker='o', label='Precision (macro)')
plt.plot(folds, fold_recalls, marker='o', label='Recall (macro)')
plt.title('Precision and Recall Across Folds')
plt.xlabel('Fold')
plt.ylabel('Score')
plt.ylim(0, 1.05)
plt.xticks(folds)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
